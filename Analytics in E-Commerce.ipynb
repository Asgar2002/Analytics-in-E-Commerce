Recency


```python
current_date=df["Order Date"].max()
recent_date=df.groupby("Customer ID")["Order Date"].max().reset_index()
print(recent_date.head())
recent_date['Recency']=(current_date-recent_date['Order Date']).dt.days
print(recent_date[['Customer ID','Recency']].head(10))
```

Frequency

```python
frequency_date=df.groupby("Customer ID")["Order Date"].nunique().reset_index()
frequency_date.columns=['Customer ID','Frequency']
print(frequency_date)
```

Monetary

```python
df['TotalPrice']=df['Quantity']*df['Sales']
```

```python
monetary_date=df.groupby("Customer ID")["TotalPrice"].sum().reset_index()
monetary_date.columns=['Customer ID','Monetary']
print(monetary_date.head(10))
```

```python
df1=pd.merge(recent_date,frequency_date)
df1=pd.merge(df1,monetary_date,on='Customer ID')
df1
```

```python
df1
```

```python
df2=df1.drop(['Order Date','Customer ID'],axis=1)
```

```python
df2
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
wcss=[]
for i in range(2,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=15)
  kmeans.fit(df2)
  wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,6))
plt.plot(range(2,11),wcss,marker='o',linestyle='--',color='red')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=6,random_state=4)
kmeans.fit(df2)
```

```python
df2['Cluster']=kmeans.labels_
df2
```

```python
for i in range(6):
  cluster_df=df2[df2['Cluster']==i]
  print(f"Cluster{i}:\n{cluster_df}\n")
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg=silhouette_score(df2,kmeans.labels_)
print('average silhoutte score:',silhouette_avg)
```

```python
cluster_count=df2['Cluster'].value_counts()
print(cluster_count)

plt.figure(figsize=(8,6))
plt.pie(cluster_count,labels=cluster_count.index,autopct='%1.1f%%',startangle=90)
plt.title('Cluster Distribution')
plt.show()
```

```python
df1['r_rank']=df1['Recency'].rank(ascending=False)#gives higher rank to the customers with lower recency
df1['f_rank']=df1['Frequency'].rank(ascending=True)#here lower the frequeny lower the rank
df1['m_rank']=df1['Monetary'].rank(ascending=True)
df1
```

```python
#normalizing the data
df1['r_rank_norm']=(df1['r_rank']/df1['r_rank'].max())*100
df1['f_rank_norm']=(df1['f_rank']/df1['f_rank'].max())*100
df1['m_rank_norm']=(df1['m_rank']/df1['m_rank'].max())*100
df1
```

```python
df1.drop(["r_rank","f_rank","m_rank"],axis=1,inplace=True)
df1
```

```python
df1['rfm_score']=0.33*df1['r_rank_norm']+0.33*df1['f_rank_norm']+0.33*df1['m_rank_norm']
df1[["Customer ID","rfm_score"]]
```

```python
df1["Customer_segment"] = np.where(df1['rfm_score'] > 55,
                                      "Top top most Customers",
                          np.where(df1['rfm_score'] > 50,
                                      "Top most Customers",
                          np.where(df1['rfm_score'] > 45,
                                      "Top Customers",
                                      (np.where(
                                        df1['rfm_score'] > 40,
                                        "High value Customer",
                                        (np.where(
                                df1['rfm_score'] > 30,
                             "Medium Value Customer",
                             np.where(df1['rfm_score'] > 1.5,
                            'Low Value Customers', 'Lost Customers'))))))))
df1[['Customer ID', 'rfm_score', 'Customer_segment']].head(20)
```

```python
a = df1["Customer_segment"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(a, labels=a.index, autopct='%1.1f%%', startangle=90)
plt.title("Customer Segmentation Based on RFM Score")
plt.show()
```
